# Seaborn

```python
# import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Important things for plotting
1. Data (df)
2. x-axis
3. y-axis
4. Type of plot
5. Grouping (hue/class) based on categorical variables
6. Size of plot
7. Color pallets of plot
8. Style of plot
9. How do you interpret

```python
# Data

df = sns.load_dataset('titanic')
```

```python
df.info()
```

```python
# Scatter plot Age and Fare

sns.scatterplot(data=df, x='age',y='fare')
plt.savefig('../outputs/sctter_plot.png', dpi=300)
```

```python
# Grouping

sns.scatterplot(data=df, x='age',y='fare', hue='sex')
plt.savefig('../outputs/sctter_plot_grouping.png', dpi=300)
```

```python
# fig size of plot
plt.figure(figsize=(8,4))
sns.scatterplot(data=df, x='age',y='fare', hue='sex')
plt.savefig('../outputs/sctter_plot_size.png', dpi=300)
```

```python
# Change the color

sns.scatterplot(data=df, x='age',y='fare', color='red')
plt.savefig('../outputs/sctter_plot_color.png', dpi=300)
```

```python
# Change color with hue

palette = {'male':'blue','female':'red'}
plt.figure(figsize=(8,4))
sns.scatterplot(data=df, x='age',y='fare', hue='sex', palette=palette)

# labels

plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter plot of Age vs Fare')

plt.savefig('../outputs/sctter_plot_labels.png', dpi=300)
```

```python
# Change color with hue

palette = {'male':'blue','female':'red'}
plt.figure(figsize=(8,4))
sns.scatterplot(data=df, x='age',y='fare', hue='sex', palette=palette)

# labels

plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter plot of Age vs Fare')

# saving a plot into 300dpi,600dpi,150dpi

plt.savefig('../outputs/sctter_plot.png', dpi=300)
```

```python
# Set Styles

sns.set_style('whitegrid')
palette = {'male':'blue','female':'red'}
plt.figure(figsize=(8,4))
sns.scatterplot(data=df, x='age',y='fare', hue='sex', palette=palette)

# labels

plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter plot of Age vs Fare')

plt.savefig('../outputs/sctter_plot_theme.png', dpi=300)
```

```python
sns.set_theme(style='darkgrid')
palette = {'male':'blue','female':'red'}
plt.figure(figsize=(8,4))
sns.scatterplot(data=df, x='age',y='fare', hue='sex', palette=palette)

# labels

plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter plot of Age vs Fare')
```

```python
# also can change font style or size

sns.set_theme(style='darkgrid', font_scale=1.5, font='Calibri')
palette = {'male':'blue','female':'red'}
plt.figure(figsize=(8,4))
sns.scatterplot(data=df, x='age',y='fare', hue='sex', palette=palette)

# labels

plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter plot of Age vs Fare')

plt.savefig('../outputs/sctter_plot_style&size.png', dpi=300)
```

### Interpetetion:
1. Scatter show relation of fare nd age of titanic passengers.
2. We have age of passengers at x-axis and fare at y-axis.
3. We further group Passengers by their Gender such as male ( green points ) and female ( red points )
4. Two male passengers paid highest fare and they migt be in first class
5. The oldest person at titanic is 80 years old and Male by gender shown on plot

```python
# box plot

sns.set_theme(style='darkgrid', font_scale=1.5, font='Calibri')
palette = {'First':'blue','Second':'red', 'Third':'green'}
plt.figure(figsize=(8,4))
sns.boxplot(data=df, x='sex',y='age', hue='class', palette=palette)

# labels

plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter plot of Age vs Fare')

plt.savefig('../outputs/box_plot.png', dpi=300)
```

```python
# Create a bar chart

palette = {'male':'blue','female':'red'}
plt.figure(figsize=(8,4))
sns.barplot(x="sex", y="fare", data=df, palette=palette)

# Show the plot
plt.show()

plt.savefig('../outputs/bar_plot_color.png', dpi=300)
```
